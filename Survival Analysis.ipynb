```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('KAts.csv', low_memory = False, parse_dates = ['Date'])
```

```python
df.head()
```

```python
# Converting the "Date" column to datetime format
df['Date']=pd.to_datetime(df['Date'].apply(lambda x: x+'-2020'))
```

# Removing "X" from column names

```python
col = pd.Series(df.columns[4:])

def remove_X(x):
    return pd.to_datetime(x[1:], dayfirst=True)

c = col.apply(lambda x: remove_X(x))
c.index = range(len(df.columns))[4:]
c  # "c" will hold the new column names
```

```python
# Appending the first four columns to "c"
c[0]=df.columns[0]
c[1]=df.columns[1]
c[2] = df.columns[2]
c[3] =df.columns[3]

# Sorting the series as per index
c.sort_index(inplace = True)

# Renaming columns
df.columns = c

del c
```

# Cured

```python
df2 = df.isin(['C']).copy()
```

```python
df2 = df2[df2==True]
df2.dropna(axis=0,how='all', inplace=True)
temp = df2.idxmax(axis=1)
```

```python
df2.head()
```

```python
df2.shape
```

```python
# Concatenating 'Age', 'Sex', 'Date'
df2[['Age', 'Sex', 'Date']] = df.loc[df2.index,['Age', 'Sex', 'Date']]
df2.head()
```

```python
temp.index = df2['Date']
temp
```

```python
cured = pd.DataFrame()
cured['Hospitalised time'] = temp - temp.index
cured.head()
```

```python
cured.reset_index(inplace=True)
cured.index = df2.index
cured[['Age', 'Sex']] = df.loc[df2.index,['Age', 'Sex']]
cured.head()
```

```python
del df2
```
